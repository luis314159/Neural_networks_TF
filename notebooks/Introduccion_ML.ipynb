# Diferencia entre programación explícita y Machine learning

Imaginate que se te pide hacer un porgrama que pueda identificar un solo digito que se un usuario escriba ¿Cómo realizarias esta tarea?
<br>
<br>


La diferencia entre programación explícita y machine learning (aprendizaje automático) se puede entender claramente a través del ejemplo de un programa diseñado para identificar un número (un dígito) escrito a mano. Vamos a comparar cómo se abordaría este problema con cada enfoque:

<br>

**<center>Programación Explícita</center>**


<br>



> En la programación explícita, un programador escribe reglas específicas y detalladas que el sistema debe seguir para identificar un dígito. Estas reglas se basan en el conocimiento previo del programador sobre cómo se ve cada número y deben cubrir todos los posibles escenarios y variaciones en la escritura a mano de los dígitos. Por ejemplo:

> El programador podría escribir un conjunto de reglas para identificar un "8" como dos círculos, uno encima del otro.
Para un "1", podría definir que es una línea vertical sin curvas significativas.
Este método requiere que el programador anticipe y codifique explícitamente cada posible variación en la apariencia de cada número, lo cual es extremadamente complejo y casi imposible de lograr de manera exhaustiva debido a la infinita variedad en la escritura a mano de las personas.

<br>


**<center>Machine Learning</center>**
<br>


> En el enfoque de machine learning, en lugar de escribir reglas específicas, el programador desarrolla un algoritmo que puede aprender patrones a partir de los datos. Para identificar dígitos, se utilizaría un conjunto de datos de entrenamiento que contiene imágenes de dígitos escritos a mano junto con sus etiquetas correctas (los números reales que representan).

> El algoritmo, como una red neuronal, aprendería a reconocer patrones en los datos, identificando características que distinguen a cada número del otro.
Después del entrenamiento, el modelo puede identificar dígitos en nuevas imágenes que nunca ha visto antes, basándose en los patrones aprendidos.
En este caso, el algoritmo aprende a identificar dígitos por sí mismo a partir de los ejemplos proporcionados, sin que el programador necesite especificar reglas detalladas para cada dígito.

> Una neurona artificial es un elemento fundamental de las redes neuronales en el aprendizaje automático, diseñada para imitar el comportamiento de una neurona biológica. Consiste en unidades de procesamiento que reciben varias señales de entrada, las combinan de alguna manera (por lo general, sumándolas y aplicándoles pesos), y luego aplican una función de activación para generar una señal de salida. La simplicidad y eficacia de las neuronas artificiales permiten construir redes complejas capaces de aprender y realizar tareas de clasificación, regresión y muchas otras en inteligencia artificial.
