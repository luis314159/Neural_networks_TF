# Diferencia entre programación explícita y Machine learning

Imaginate que se te pide hacer un porgrama que pueda identificar un solo digito que se un usuario escriba ¿Cómo realizarias esta tarea?
<br>
<br>


La diferencia entre programación explícita y machine learning (aprendizaje automático) se puede entender claramente a través del ejemplo de un programa diseñado para identificar un número (un dígito) escrito a mano. Vamos a comparar cómo se abordaría este problema con cada enfoque:

<br>

**<center>Programación Explícita</center>**


<br>



> En la programación explícita, un programador escribe reglas específicas y detalladas que el sistema debe seguir para identificar un dígito. Estas reglas se basan en el conocimiento previo del programador sobre cómo se ve cada número y deben cubrir todos los posibles escenarios y variaciones en la escritura a mano de los dígitos. Por ejemplo:

> El programador podría escribir un conjunto de reglas para identificar un "8" como dos círculos, uno encima del otro.
Para un "1", podría definir que es una línea vertical sin curvas significativas.
Este método requiere que el programador anticipe y codifique explícitamente cada posible variación en la apariencia de cada número, lo cual es extremadamente complejo y casi imposible de lograr de manera exhaustiva debido a la infinita variedad en la escritura a mano de las personas.

<br>


**<center>Machine Learning</center>**
<br>


> En el enfoque de machine learning, en lugar de escribir reglas específicas, el programador desarrolla un algoritmo que puede aprender patrones a partir de los datos. Para identificar dígitos, se utilizaría un conjunto de datos de entrenamiento que contiene imágenes de dígitos escritos a mano junto con sus etiquetas correctas (los números reales que representan).

> El algoritmo, como una red neuronal, aprendería a reconocer patrones en los datos, identificando características que distinguen a cada número del otro.
Después del entrenamiento, el modelo puede identificar dígitos en nuevas imágenes que nunca ha visto antes, basándose en los patrones aprendidos.
En este caso, el algoritmo aprende a identificar dígitos por sí mismo a partir de los ejemplos proporcionados, sin que el programador necesite especificar reglas detalladas para cada dígito.

# Cambiando las etiquetas
Como las etiquetas son valores de "String" es decir texto, los modelos de machine learning no entienden esto, ellos entienden de números, así que cambiaremos los valores de 'Iris-versicolor' por 1 e "Iris-setosa" por 0.

```python
def versicolor_to_one(x):
  if x == 'Iris-versicolor':
    return 1
  else:
    return 0
```

```python
y = y.apply(versicolor_to_one)
```

```python
y.head() # mostrando primeros 5 valores
```

### Ahora a entrenar nuestro modelo..... Un perceptron
Hemos cambiado las etiquetas a 0 y 1 porque usaremos un perceptron, es decir una neurona artificial, la neurona encendera si se trata de 'Iris-versicolor' y permanecera apagada si se trata de 'Iris-setosa'
Le preguntaremos al chatGPT :)


```python
# Chat dijo esto (si lo dice chatGPT debe estar bien :D)

import tensorflow as tf


# Definir el modelo del perceptrón
modelo = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='sigmoid', input_shape=(2,)),
])

# Especificar el learning rate deseado
learning_rate = 0.01  # Cambia el valor del learning rate según tus necesidades

# Crear el optimizador personalizado con el learning rate especificado
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

# Compilar el modelo con el optimizador y la función de pérdida
modelo.compile(optimizer=optimizer, loss = 'binary_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
modelo.fit(X, y, epochs=100, batch_size = 32)
```

### Qué onda con el Perceptron?


<img src = "https://miro.medium.com/v2/resize:fit:639/1*_Epn1FopggsgvwgyDA4o8w.png">

Este mecanismo emula la operación de una neurona biológica, asignando un "peso" específico a cada señal de entrada. Cada una de estas entradas se multiplica por su peso respectivo, y el resultado de estas multiplicaciones se suma. Posteriormente, se aplica una función de activación a esta suma acumulada: si el valor resultante excede el umbral de 0, la neurona se considera activada, modificando su estado de salida en consecuencia.

# Cómo aprende el modelo?

<img src = "https://img.freepik.com/fotos-premium/montanas-region-quindio-colombia_422213-263.jpg?w=2000">


Montañas por qué wtf?

Por el descenso del gradiente............


El modelo aspira a alcanzar el punto más bajo del terreno, operando a ciegas, sin capacidad para ver. Su único indicio de progreso es la sensación de ascenso o descenso tras cada movimiento realizado. La magnitud de estos movimientos, ya sean amplios o limitados, está dictada por el "tasa de aprendizaje" (LR). Este parámetro define la escala de los pasos que el modelo puede tomar, influenciando directamente su capacidad para navegar por el paisaje hacia su objetivo.


[Herramienta interactiva](http://www.benfrederickson.com/numerical-optimization/)

### Cambiar de string a One hot encoding

 Debido a qué tenemos valores de string y el modelo necesita saber qué neurona se activara con cada planta, necesitamos cambiar de string a categorico en One hot encoding.

```python
# Función con la lógica para cambniar a one_hot encoding
def encode_species(x):
    if x == 'Iris-versicolor':
        return [0, 0, 1]
    elif x == 'Iris-setosa':
        return [0, 1, 0]
    else:  # Se asume que la única otra opción es 'Iris-virginica'
        return [1, 0, 0]
```

```python
# Aplicando la función definida
y2_encoded = y2.apply(encode_species)
```

```python
#Convirtiendo y2 a una lista y luego a un numpy array:
import numpy as np
y2_encoded_array = np.array(y2_encoded.tolist())
y2_encoded_array
```

```python
y2_encoded_array.shape
```

### Crear nuevo modelo



```python
# Chat dijo esto (si lo dice chatGPT debe estar bien):

import tensorflow as tf

# Definir el modelo del perceptrón
modelo2 = tf.keras.Sequential([
    tf.keras.layers.Dense(3, activation='sigmoid', input_shape=(2,)),
])

# Especificar el learning rate deseado
learning_rate = 0.1  # Cambia el valor del learning rate según tus necesidades

# Crear el optimizador personalizado con el learning rate especificado
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

# Compilar el modelo con el optimizador y la función de pérdida
modelo2.compile(optimizer=optimizer, loss = 'categorical_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
modelo2.fit(X2, y2_encoded_array, epochs=100)
```

### Volviendo a entrenar el modelo con un LR más chico


```python

# Especificar el learning rate deseado
learning_rate = 0.001  # Cambia el valor del learning rate según tus necesidades

# Crear el optimizador personalizado con el learning rate especificado
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

# Compilar el modelo con el optimizador y la función de pérdida
modelo2.compile(optimizer=optimizer, loss = 'categorical_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
modelo2.fit(X2, y2_encoded_array, epochs=100)
```

## Overfitting and Evaluation
Es importante que el modelo generalize apartir de los datos, es decir que pueda clasificar correctamente datos afuera de los datos de entrenamiento.

El modelo puede presentar overfitting (Sobreajuste) al ser demasiado complejo para la tarea y puede presentar underfitting (subajuste) cuando no es lo suficientemente complejo, es decir no es capaz de aprender.


Para verificar que no tiene sobre ajuste podemos probar el modelo con otros datos con los que no sea haya entrenado.

<img src = "https://www.mathworks.com/discovery/overfitting/_jcr_content/mainParsys/image.adapt.full.medium.svg/1686825007300.svg">
